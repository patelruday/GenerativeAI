```python
#import streamlit as st
from transformers import BartForConditionalGeneration, BartTokenizer
#from langchain.text_splitter import RecursiveCharacterTextSplitter

# Load pretrained BART models
model_name = "facebook/bart-large-cnn"
tokenizer = BartTokenizer.from_pretrained(model_name)
model = BartForConditionalGeneration.from_pretrained(model_name)

# define a function to make chunks from tokens
def make_chunks(text, max_tokens=80, overlap_tokens=10):
    inputs = tokenizer.encode("summarize"+text,return_tensors="pt",padding=True)
    tokens = inputs[0]
    chunks = []
    print(f"no of tokens{len(tokens)}")

    # Iteratively create overlapping chunks
    for i in range(0, len(tokens), max_tokens - overlap_tokens):
        chunk = tokens[i:i + max_tokens]
        chunks.append(chunk)
        print(f"Chunk {len(chunks) - 1} size: {chunk.shape[0]} tokens")

    print(f"Total number of chunks: {len(chunks)}")
    return chunks




# Function to summarize each chunk
def summarize_chunk(chunk, max_length=100, min_length=50, num_beams=4):
    try:
        summary_ids = model.generate(chunk.unsqueeze(0), max_length=max_length, min_length=min_length, num_beams=num_beams, early_stopping=True)
        return tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    except Exception as e:
        print(f"Error during summarization: {e}")
        return ""

text="""Once upon a time, in a small village, there lived a hardworking man named Ramlal with his family. Ramlal was a farmer, known for his dedication and honesty. His wife, Sheela, was a gentle and wise woman who took great care of their home and children. They had four children: Raj, Kumar, Shyam, and Babita.

Raj, the eldest, was responsible and helped his father with the farm. He was strong, always ready to lend a hand, and had a passion for learning new ways to improve their crops. Kumar, the second eldest, was clever and quick-witted. Though not as strong as Raj, he had a way with numbers and often helped his mother Sheela manage the family’s finances.

Shyam, the third child, was adventurous and curious. He spent his days exploring the village and the nearby forests, often returning home with exciting stories and small treasures he found. His imagination was boundless, and he dreamt of seeing the world beyond the village. Babita, the youngest, was gentle and kind-hearted. She loved animals and often spent her time helping stray animals or tending to the family's small garden.

One year, the village faced a severe drought, and Ramlal’s crops were at risk. Worried, he gathered his family and shared his concerns. Raj offered to work extra hours in the field, while Kumar suggested they budget their savings wisely. Shyam, inspired by his love for exploration, came up with an idea. "Father," he said, "I’ve heard of a river not too far from here. Perhaps we could build a canal to bring water to our fields."

Ramlal was hesitant, but Sheela encouraged them to work together on Shyam’s plan. Over the next few weeks, the whole family pitched in. Raj and Ramlal dug the canal, Kumar calculated the materials they would need, and Shyam guided them through the safest route to bring the water. Babita brought them food and kept everyone's spirits high with her sweet laughter.

Finally, after weeks of hard work, water flowed into their fields. The crops thrived, and the family's efforts saved their farm. The village hailed Ramlal’s family as heroes, inspired by their unity, love, and determination. They proved that, together, a family could overcome any challenge.

And from that day on, Ramlal’s family became an example of teamwork and resilience for everyone in the village.
Once upon a time, in a small village, there lived a hardworking man named Ramlal with his family. Ramlal was a farmer, known for his dedication and honesty. His wife, Sheela, was a gentle and wise woman who took great care of their home and children. They had four children: Raj, Kumar, Shyam, and Babita.

Raj, the eldest, was responsible and helped his father with the farm. He was strong, always ready to lend a hand, and had a passion for learning new ways to improve their crops. Kumar, the second eldest, was clever and quick-witted. Though not as strong as Raj, he had a way with numbers and often helped his mother Sheela manage the family’s finances.

Shyam, the third child, was adventurous and curious. He spent his days exploring the village and the nearby forests, often returning home with exciting stories and small treasures he found. His imagination was boundless, and he dreamt of seeing the world beyond the village. Babita, the youngest, was gentle and kind-hearted. She loved animals and often spent her time helping stray animals or tending to the family's small garden.

One year, the village faced a severe drought, and Ramlal’s crops were at risk. Worried, he gathered his family and shared his concerns. Raj offered to work extra hours in the field, while Kumar suggested they budget their savings wisely. Shyam, inspired by his love for exploration, came up with an idea. "Father," he said, "I’ve heard of a river not too far from here. Perhaps we could build a canal to bring water to our fields."

Ramlal was hesitant, but Sheela encouraged them to work together on Shyam’s plan. Over the next few weeks, the whole family pitched in. Raj and Ramlal dug the canal, Kumar calculated the materials they would need, and Shyam guided them through the safest route to bring the water. Babita brought them food and kept everyone's spirits high with her sweet laughter.

Finally, after weeks of hard work, water flowed into their fields. The crops thrived, and the family's efforts saved their farm. The village hailed Ramlal’s family as heroes, inspired by their unity, love, and determination. They proved that, together, a family could overcome any challenge.

And from that day on, Ramlal’s family became an example of teamwork and resilience for everyone in the village.
"""


chunks = make_chunks(text)
summaries = [summarize_chunk(chunk) for chunk in chunks]
final_summary = " ".join(summaries)
print("Generated summary",final_summary)
#for i,chunk in enumerate(chunks):
   # print(f"chunk{i+1}/n{chunk} ")
```
